```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the csv file using Pandas
df = pd.read_csv('ElectricCarData_Clean.csv')
```

```python
# Display the first 50 rows
df.head(50)
```

```python
# Convert TopSpeed_KmH column to MpH and change the column name
convert_to_mph = 0.621371
df['TopSpeed_KmH'] = (df['TopSpeed_KmH'] * convert_to_mph).round()
df = df.rename(columns={'TopSpeed_KmH': 'Top Speed MPH'})
# Display the new dataframe
df.head(10)
```

```python
convert_to_miles = 0.621371
df['Range_Km'] = (df['Range_Km'] * convert_to_miles).round()
df = df.rename(columns={'Range_Km': 'Range (miles)'})
df.head()
```

```python
# Update the currency columns to USD and change the name of the column
currency_update = 1.08
df['PriceEuro'] = (df['PriceEuro'] * currency_update)
df = df.rename(columns={'PriceEuro': 'Price (USD)'})
df.head(20)
```

```python
# Update the Efficiency_WhKm coloumn to an Imperial measurement and change the name of the column
df['Efficiency_WhKm'] = (df['Efficiency_WhKm'] * convert_to_miles).round()
df = df.rename(columns={'Efficiency_WhKm': 'Efficiency WhM'})
df.head(20)
```

```python
# Convert all values in FastCharge_KmH from '-' to NaN 
df['FastCharge_KmH'] = df['FastCharge_KmH'].replace('-', np.nan)

# Replacing Null values in FastCharge_KmH column to 0
df['FastCharge_KmH'].fillna(1, inplace= True)



df.tail(50)
```

```python
# Delete the FastCharge_KmH column
del df['FastCharge_KmH']
df
```

```python
# Assign df variable to a new one to assure it has been cleaned
cleaned_df = df
cleaned_df.head()
```

```python
# Sort the cleaned dataframe by Brands
df_sorted = cleaned_df.sort_values(by='Brand')
df_sorted
```

```python
# Find the average price of an EV per car brand and show the result
average_price_per_brand = df_sorted.groupby("Brand")['Price (USD)'].mean().round()
average_price_per_brand
```

```python
# Find the avereage electric range of an EV by car brand and show the result
average_range_per_brand = df.groupby("Brand")['Range (miles)'].mean().round()
print(average_range_per_brand)
```

```python
# Find the average price of an EV by Brand
avg_price = df_sorted.groupby('Brand')['Price (USD)'].mean().reset_index()
avg_price = avg_price.sort_values(by='Price (USD)')

#Graph using new dependency Seaborn
sns.barplot(x='Brand', y='Price (USD)', data= avg_price, palette='colorblind') # Palette is for you, Phil!

plt.xlabel('Car Brands')
plt.ylabel("Average Price (USD)")
plt.title('Average Price of Electric Vehichles by Brand')
plt.tight_layout()
plt.xticks(rotation=90)
plt.savefig('avg_price.png', bbox_inches= 'tight')
plt.show()
```

```python
# Find the average range of an EV by Brand
avg_range = df_sorted.groupby('Brand')['Range (miles)'].mean().reset_index()
avg_range = avg_range.sort_values(by='Range (miles)')

# Graph using new dependency Seaborn
sns.barplot(x='Brand', y='Range (miles)', data= avg_range, palette='colorblind') # Palette is for you, Phil!

plt.xlabel('Car Brands')
plt.ylabel("Average Range (Miles)")
plt.title('Driveable Distance On Full Battery')
plt.tight_layout()
plt.xticks(rotation=90)
plt.savefig('avg_range.png', bbox_inches='tight')
plt.show()
```

```python
# Find all of the prices of Tesla Car models
tesla_df = df[df['Brand'] == 'Tesla ']
tesla_df
```

```python
# Drop any duplicate models to ensure a clean graph
unique_models_df = tesla_df.drop_duplicates(subset=['Model'])
unique_models_df
```

```python
# Graph the Price of Tesla Models using Seaborn
sns.barplot(x='Model', y='Price (USD)', data=unique_models_df, palette='colorblind')
plt.xlabel('Models')
plt.ylabel('Price (USD)')
plt.title("Prices of Tesla Models")
plt.tight_layout()
plt.xticks(rotation = 90)
plt.savefig('tesla_model_prices.png', bbox_inches= 'tight')
plt.show()
```
